## Imputar datos

```python
# Reemplazar los espacios en blanco (' ') por NaN
df.replace(' ', np.nan, inplace=True)
```

```python
df.shape
```

```python
# Proporcion de valores nulos por columna
df.isnull().sum()/df.shape[0]
```

```python
# Ver valores unicos por cada columna
for columna in df.columns:
    valores_unicos = df[columna].unique()
    print(f"Valores únicos en la columna '{columna}':")
    print(valores_unicos)
    print()
```

**TIPO_VIVIENDA**

```python
# Reemplazar los valores nulos por la moda
df['TIPO_VIVIENDA'].fillna(df['TIPO_VIVIENDA'].mode().iloc[0], inplace = True)
```

```python
df.head()
```

**MATERIAL_PAREDES**

```python
# Reemplazar los valores nulos por la moda
df['MATERIAL_PAREDES'].fillna(df['MATERIAL_PAREDES'].mode().iloc[0], inplace = True)
```

**NUM_HABITACIONES**

```python
df['NUM_HABITACIONES'].dtypes
```

```python
# Cambiar el tipo de dato del NUM_HABITACIONES de object a int64

# Convierte los valores en números (incluyendo los NaN)
df['NUM_HABITACIONES'] = pd.to_numeric(df['NUM_HABITACIONES'], errors='coerce') # Los valores no válidos se
                                                                                # convertirán en valores nulos

# Calcula la media de los valores numéricos
media = df['NUM_HABITACIONES'].mean()

# Llena los valores nulos con la media
df['NUM_HABITACIONES'].fillna(media, inplace=True)
```

```python
# Cambiar tipo de dato a int64
df['NUM_HABITACIONES'] = df['NUM_HABITACIONES'].astype('int64')
```

```python
df['NUM_HABITACIONES'].dtypes
```

**ESTADO**

```python
# Reemplazar los valores nulos por la moda
df['ESTADO'].fillna(df['ESTADO'].mode().iloc[0], inplace = True)
```

**MONTO_MENSUAL_ALQUILER_COMPRA**

```python
# Como el 92.7021% es nulo se dropea la variable
df = df.drop('MONTO_MENSUAL_ALQUILER_COMPRA', axis=1)
```

**AGUA**

```python
# Reemplazar los valores nulos por la moda
df['AGUA'].fillna(df['AGUA'].mode().iloc[0], inplace = True)
```

```python
# Reemplazar los valores nulos por la moda
df['ACCESO_AGUA'].fillna(df['ACCESO_AGUA'].mode().iloc[0], inplace = True)
```

**SIN_SERVICIOS_HIGIENICOS**

```python
# En el caso de SIN_SERVICIOS_HIGIENICOS, reemplazmos los valores nulos con una nueva categoria DESCONOCIDO (D)
df.loc[df.SIN_SERVICIOS_HIGIENICOS.isnull(), 'SIN_SERVICIOS_HIGIENICOS'] = 'D'
```

**NIÑOS_NO_COLEGIO**

```python
# En el caso de NIÑOS_NO_COLEGIO, reemplazmos los valores nulos con una nueva categoria DESCONOCIDO (D)
df.loc[df.NIÑOS_NO_COLEGIO.isnull(), 'NIÑOS_NO_COLEGIO'] = 'D'
```

**TIENE_TELEFONO, TIENE_CELULAR, TIENE_TV y TIENE_INTERNET**

```python
# Se reemplaza los valores nulos por un D de desconocido
df = df.fillna(value = {'TIENE_TELEFONO':'D', 'TIENE_CELULAR':'D', 'TIENE_TV':'D', 'TIENE_INTERNET':'D'})
```
