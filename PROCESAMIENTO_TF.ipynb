```python
# Calcular el precio promedio por tipo de vivienda
avg_price_by_vivienda = df.groupby('TIPO_VIVIENDA')['PRECIO'].mean().sort_values(ascending=False)

# Filtrar el DataFrame original para incluir solo esos tipos de vivienda
filtered_df = df[df['TIPO_VIVIENDA'].isin(avg_price_by_vivienda.index)]

plt.figure(figsize=(10,6))
sns.barplot(data=filtered_df, x='TIPO_VIVIENDA', y='PRECIO', estimator=np.mean, order=avg_price_by_vivienda.index)
plt.title('Precio Promedio por Tipo de Vivienda')
plt.xlabel('Tipo de Vivienda')
plt.ylabel('Precio Promedio')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mejor visualización
plt.show()
```

```python
# Calcular el precio promedio por tipo de vivienda
avg_price_by_vivienda = df.groupby('TIPO_VIVIENDA')['PRECIO'].mean().sort_values(ascending=False)

# Seleccionar los 7 tipos de vivienda con el precio promedio más alto
top_7_avg_price = avg_price_by_vivienda.head(7)

print(top_7_avg_price)
```

```python
df['REGION'] = np.where(df['DOMINIO'].str.contains('Costa'), 'Costa',
                          np.where(df['DOMINIO'].str.contains('Sierra'), 'Sierra',
                                   np.where(df['DOMINIO'].str.contains('Selva'), 'Selva',
                                            'Lima')))
```

```python
# Filtrar las viviendas que tienen Celular, Internet, TV y Teléfono para cada región
viviendas_con_celular = df[df['TIENE_CELULAR'] == 1].groupby('REGION').size()
viviendas_con_internet = df[df['TIENE_INTERNET'] == 1].groupby('REGION').size()
viviendas_con_tv = df[df['TIENE_TV'] == 1].groupby('REGION').size()
viviendas_con_telefono = df[df['TIENE_TELEFONO'] == 1].groupby('REGION').size()

# Crear un DataFrame con los datos
servicios = pd.DataFrame({
    'Tiene Celular': viviendas_con_celular,
    'Tiene Internet': viviendas_con_internet,
    'Tiene TV': viviendas_con_tv,
    'Tiene Teléfono': viviendas_con_telefono
})

servicios
```

```python
# Asumiendo que ya tienes el DataFrame servicio cargado
fig, axs = plt.subplots(2, 2, figsize=(14, 14))

# Títulos para los gráficos
titles = ['Tiene Celular', 'Tiene Internet', 'Tiene TV', 'Tiene Teléfono']

for ax, title in zip(axs.ravel(), titles):
    ax.pie(servicios[title], labels=servicios.index, autopct='%1.1f%%', startangle=90)
    ax.set_title(title)

plt.tight_layout()
plt.show()
```
